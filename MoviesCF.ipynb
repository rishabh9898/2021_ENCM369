```python
import pandas as pd
```

```python
ratings_csv = pd.read_csv('/content/drive/MyDrive/Collaborative Filtering Dataset/dataset/ratings.csv')
movies = pd.read_csv('/content/drive/MyDrive/Collaborative Filtering Dataset/dataset/movies.csv')

ratings = pd.merge(movies,ratings_csv).drop(['genres','timestamp'],axis = 1)
user_ratings = ratings.pivot_table(index = ['userId'], columns = ['title'] , values = 'rating')
user_ratings = user_ratings.dropna(thresh = 10, axis = 1).fillna(0)
item_similiarity_df = user_ratings.corr(method = 'pearson') 


def get_similiar_movies(movie_name,user_rating):
    similiar_score = item_similiarity_df[movie_name]*(user_rating-2.5)
    similiar_score = similiar_score.sort_values(ascending = False)

    return similiar_score


movie_input = str(input("Enter a movie you like and we will recommend something for you : ? " ))
rating_input = int(input("Enter the rating you would give to this film "))


action_lover = [(movie_input,rating_input)]

similiar_movies = pd.DataFrame()

for movie,rating in action_lover:
    similiar_movies = similiar_movies.append(get_similiar_movies(movie,rating),ignore_index= True)


similiar_movies.head()
similiar_movies.sum().sort_values(ascending = False)[:5]
```
